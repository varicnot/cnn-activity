### Note: Augmentation was applied only to the training dataset, not the validation dataset.

```python
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# We implemented augmentation using TensorFlow Keras preprocessing layers
base = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)
```

Transformations Used
1. RandomRotation (±12%)
- Rotates images randomly within ±12% of a full rotation.

2. RandomTranslation (15% height and width)
- Shifts images horizontally and vertically up to 15%.

3. RandomZoom (±15%)
- Randomly zooms in or out by up to 15%.

```python
data_aug = tf.keras.Sequential([
    tf.keras.layers.RandomRotation(0.08),
    tf.keras.layers.RandomTranslation(0.10, 0.10),
    tf.keras.layers.RandomZoom(0.10),
    tf.keras.layers.RandomContrast(0.10),
])
```

```python
DATA_DIR = "LT3_digits/LT3_digits"
IMG_SIZE = (224, 224)
BATCH = 16
SEED = 42

train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    color_mode="rgb",
    label_mode="int"
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    color_mode="rgb",
    label_mode="int"
)

class_names = train_ds.class_names
print(class_names)
```

```python
IMG_SIZE = (224, 224)
AUTOTUNE = tf.data.AUTOTUNE

# Augmentation (digits-safe)
data_aug = tf.keras.Sequential([
    tf.keras.layers.RandomRotation(0.12),
    tf.keras.layers.RandomTranslation(0.15, 0.15),
    tf.keras.layers.RandomZoom(0.15),
])

# MobileNetV2 preprocessing
preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

train_ds2 = train_ds.map(lambda x, y: (preprocess(data_aug(x, training=True)), y),
                         num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)

val_ds2   = val_ds.map(lambda x, y: (preprocess(x), y),
                       num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)
```

```python
base = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)
base.trainable = False

model = tf.keras.Sequential([
    base,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(10, activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    train_ds2,
    validation_data=val_ds2,
    epochs=20,
    callbacks=[early_stop]
)
```

```python
model.evaluate(val_ds2)
```

```python
# Fine-tuning stage

base.trainable = True

# Freeze early layers, unfreeze last 30
for layer in base.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),  # lower LR for fine-tuning
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_finetune = model.fit(
    train_ds2,
    validation_data=val_ds2,
    epochs=10
)
```

```python
y_true = []
y_pred = []

for images, labels in val_ds2:
    preds = model.predict(images, verbose=0)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names)

plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Validation)")
plt.show()
```
